```python
data = {"name" : ["sponge bob", "Tom cruise", "John wick"], 
       "age" : [10, 23, 43 ],
        "email" : ["sponge@gmail.com", "tom@gmail.com", "john@gmail.com" ]
       }
df = pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])

df["job"] = ["Act as cock", "Act as CIA", "Act as killer"]

new_dt = pd.DataFrame([{"name" : "Sandy", "age" : 16, "email" : "sandy@gmail.com", "job" : "astronut"}, 
                      {"name" : "airmjay", "age" : 10, "email" : "airmjay@gmail.com", "job" : "programmer"}], 
                      index=["Employee 4", "Employee 5"])
df = pd.concat([df, new_dt])
print(df.loc["Employee 1"])
```

```python
df = pd.read_csv("pre.csv", index_col="UID")
print(df.to_string())
```

```python
#take user input
userPrompt = input("Enter the UID: ")
try:
    print(df.loc[userPrompt])
except KeyError:
    print(f"{userPrompt} not found")
```

```python
#filter
president = df[df["Position"] == 'President']
print(president.to_string())
```

```python
#filter pres and vice
pv = df[(df['Position'] == 'President') | (df['Position'] == 'Vice President')]
print(pv.to_string())
```

```python
#filter pres and vice under tinubu administration 
pv = df[((df['Position'] == 'President') | (df['Position'] == 'Vice President')) & (df['AdminName'] == 'Tinubu Admin')]
print(pv.to_string())
```

```python
#aggregate function
print(df.count())
```

```python
#group by
group = df.groupby("Position")
print(group["Position"].count())
```

```python
#data cleaning 
df = df.drop(columns=['Category'])
print(df)
```

```python
df['AdminID'] = df["AdminID"].replace({'BAT-2023' : 'CITY-BOY'})
print(df)
```

```python
#lowercase
df["AdminName"] = df["AdminName"].str.lower()
print(df)
```
