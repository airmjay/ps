## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#with python list  
x = [2012,2013,2014,2015,2016,2017,2018,2019,2020]
y = [12,43,67,39,69,38,89,39,100]

plt.plot(x,y)
plt.show()
```

```python
#with numpy array
import numpy as np
x = np.array([2012,2013,2014,2015,2016,2017,2018,2019,2020])
y = np.array([12,43,67,39,69,38,89,39,100])

plt.plot(x,y)
plt.show()
```

```python
#customize chart style and add line 
y1 = [1,12,6,11,78,54,100,39,45]
y2 = [10,15,18,19,40,78,86,90,120]
style = dict(marker='.',ms=12,mfc='green',mec='green',ls='dotted',lw=2)
plt.title("Trends",fontsize=15, family='sans-serif',fontweight='light',color='blue')
plt.xlabel("Years",fontsize=12, family='sans-serif',fontweight='light',color='blue')
plt.ylabel("Sales",fontsize=12, family='sans-serif',fontweight='light',color='blue')
plt.grid()
plt.plot(x,y, **style )
plt.plot(x,y1, **style, color='red' )
plt.plot(x,y2, **style, color='black' )
plt.show()
```

```python
#customize bar charts
month = np.array(['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'])
sales = np.array([100,213,430,212,675,786,1000,456,210,786,342,894])
sumAmount = np.sum(sales)
plt.title("AirmJay Sales", fontsize='17', family='monospace', fontweight='bold', color='steelblue')
plt.xlabel("Months", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.ylabel("Amount", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.bar(month,sales)
plt.show()
```

```python
#custimize pie chart
#customize bar charts
month = np.array(['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'])
sales = np.array([100,213,430,212,675,786,1000,456,210,786,342,894])
sumAmount = np.sum(sales)
plt.title("AirmJay Sales", fontsize='17', family='monospace', fontweight='bold', color='steelblue')
plt.xlabel("Months", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.ylabel("Amount", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.pie(sales, labels=month, autopct='%1.1f%%')
plt.show()
```

```python
#scatter graphs
x = [0,1,4,10]
y = [20,50,70,98]
x1 = [0,1,4,10]
y1 = [10,40,60,100]
plt.grid()
plt.scatter(x1,y1, label='Class B')
plt.scatter(x,y, label='Class A')
plt.legend()
```

```python
#histogram
scores = np.random.normal(loc=80,scale=10,size=100)
scores = np.clip(scores,0,100)
plt.hist(scores,edgecolor="black")
plt.grid()
plt.show()
```

```python
#subplots
figure, axes = plt.subplots(2,2)
x1 = [10,21,45,35,78]
y1 = [23,65,78,34,67]
y2 = [32,45,76,78,65]
y3 = [23,56,76,45,87]
y4 = [54,76,23,43,54]
axes[0,0].plot(x1,y1)
axes[0,1].plot(x1,y2)
axes[1,0].plot(x1,y3)
axes[1,1].plot(x1,y4)
axes[0,0].set_title('plot 0,0')
axes[0,1].set_title('plot 0,1')
axes[1,0].set_title('plot 1,0')
axes[1,1].set_title('plot 1,1')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Load data
df = pd.read_csv('customers.csv')

# 2. Select columns
data = df[['Index', 'First Name', 'Last Name', 'Company', 'City']]

# 3. CLEANING (Crucial): Remove leading/trailing spaces from names
data['First Name'] = data['First Name'].astype(str).str.strip()

# 4. Count occurrences
counts = data['First Name'].value_counts()

# 5. Get names that appear 6 or more times ( > 6 )
names_to_keep = counts[counts >= 5].index

# 6. Filter the original dataframe
df_filtered = data[data['First Name'].isin(names_to_keep)]
plot_counts = df_filtered['First Name'].value_counts()

plt.barh(plot_counts.index,plot_counts)
```
