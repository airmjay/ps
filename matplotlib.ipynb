```python
#customize chart style and add line 
y1 = [1,12,6,11,78,54,100,39,45]
y2 = [10,15,18,19,40,78,86,90,120]
style = dict(marker='.',ms=12,mfc='green',mec='green',ls='dotted',lw=2)
plt.title("Trends",fontsize=15, family='sans-serif',fontweight='light',color='blue')
plt.xlabel("Years",fontsize=12, family='sans-serif',fontweight='light',color='blue')
plt.ylabel("Sales",fontsize=12, family='sans-serif',fontweight='light',color='blue')
plt.grid()
plt.plot(x,y, **style )
plt.plot(x,y1, **style, color='red' )
plt.plot(x,y2, **style, color='black' )
plt.show()
```

```python
#customize bar charts
month = np.array(['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'])
sales = np.array([100,213,430,212,675,786,1000,456,210,786,342,894])
sumAmount = np.sum(sales)
plt.title("AirmJay Sales", fontsize='17', family='monospace', fontweight='bold', color='steelblue')
plt.xlabel("Months", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.ylabel("Amount", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.bar(month,sales)
plt.show()
```

```python
#custimize pie chart
#customize bar charts
month = np.array(['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'])
sales = np.array([100,213,430,212,675,786,1000,456,210,786,342,894])
sumAmount = np.sum(sales)
plt.title("AirmJay Sales", fontsize='17', family='monospace', fontweight='bold', color='steelblue')
plt.xlabel("Months", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.ylabel("Amount", fontsize='12', family='monospace', fontweight='bold', color='steelblue')
plt.pie(sales, labels=month, autopct='1')
plt.show()
```
